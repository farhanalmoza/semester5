```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
ganja = cv2.imread('ganja.jpg')
jarak_yodium = cv2.imread('jarak_yodium.jpg')
pepaya = cv2.imread('pepaya.jpg')
singkong = cv2.imread('singkong.jpg')
```

```python
# grayscale image
ganja = cv2.cvtColor(ganja, cv2.COLOR_BGR2GRAY)
jarak_yodium = cv2.cvtColor(jarak_yodium, cv2.COLOR_BGR2GRAY)
pepaya = cv2.cvtColor(pepaya, cv2.COLOR_BGR2GRAY)
singkong = cv2.cvtColor(singkong, cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(10, 10))
plt.subplot(221)
plt.imshow(ganja)
plt.subplot(222)
plt.imshow(jarak_yodium)
plt.subplot(223)
plt.imshow(pepaya)
plt.subplot(224)
plt.imshow(singkong)
plt.show()
```

```python
# shwo histogram
plt.figure(figsize=(10, 10))
plt.subplot(221)
plt.hist(ganja.ravel(), 256, [0, 256])
plt.subplot(222)
plt.hist(jarak_yodium.ravel(), 256, [0, 256])
plt.subplot(223)
plt.hist(pepaya.ravel(), 256, [0, 256])
plt.subplot(224)
plt.hist(singkong.ravel(), 256, [0, 256])
plt.show()
```
