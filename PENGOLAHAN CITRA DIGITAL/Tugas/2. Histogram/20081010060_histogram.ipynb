```python
import numpy as np
import skimage.color
import skimage.io
# %matplotlib widget

image = skimage.io.imread(fname="gambar.jpg", as_gray=True)

fig, axes = plt.subplots()
plt.imshow(image, cmap="gray")
```

```python
# create histogram
histogram, bin_edges = np.histogram(image, bins=256, range=(0, 1))
```

```python
# config and draw histogram
plt.figure()
plt.title("Histogram")
plt.xlabel("Grayscale value")
plt.ylabel("Pixels count")
plt.xlim([0.0, 1.0])

plt.plot(bin_edges[0:-1], histogram)
```
